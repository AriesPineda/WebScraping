```python
#selenium Automation and web scraping

#Remember to change all elements
#Remember to change the csv name
#Espescially the div and "Next"

import os
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup

# Initialize the web driver (replace 'chromedriver.exe' with the path to your web driver)
driver = webdriver.Chrome()

# URL of the webpage to start from
url = input('Enter the URL: ')
driver.get(url)

# Wait for 10 seconds (adjust as needed)
time.sleep(2)

# Number of times to click the "Next" button and parse HTML
num_clicks = int(input('How many pages:'))

time.sleep(2)

# Define the folder where HTML files will be saved
output_folder = 'C://Users//ADMIN//Documents//Web Scraping//Output_Folder'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Initialize a list to store data from all HTML files
csv_data = []

for i in range(num_clicks):
    # Wait for the page to load (you may need to adjust the wait time)
    time.sleep(5)

    # Get the current page's HTML content
    html_content = driver.page_source

    # Save the HTML content to a file with a unique filename
    filename = os.path.join(output_folder, f'page_{i + 1}.html')
    with open(filename, 'w', encoding='utf-8') as html_file:
        html_file.write(html_content)

    print(f'Page {i + 1} saved as {filename} successfully.')

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all div elements with class="" and extract data
    divs = soup.find_all('div', class_='sg-col-inner')
    for div in divs:
        # Initialize variables for link, title, rating, and price
        link = div.find('a', class_='a-link-normal s-no-outline')
        link = link.get('href') if link else ''
        title = div.find('span', class_='a-size-base-plus a-color-base a-text-normal')
        title = title.text.strip() if title else ''
        rating = div.find('span', class_='a-icon-alt')
        rating = rating.text.strip() if rating else ''
        num_rating = div.find('span', class_='a-size-base s-underline-text')
        num_rating = num_rating.text.strip() if num_rating else ''        
        price = div.find('span', class_='a-price-whole')
        price = price.text.strip() if price else ''

        # Add the data to the list
        csv_data.append([link, title, rating,num_rating, price])

    # Find and click the "Next" button by its link text
    try:
        next_button = driver.find_element(By.LINK_TEXT, 'Next')
        next_button.click()
    except NoSuchElementException:
        # If "Next" button is not found, break the loop
        print('No more pages available.')
        break

# Close the web driver

driver.quit()

# Write the data to a data.csvn file
with open('data.csv', 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    # Write header
    csv_writer.writerow(['Link', 'Title', 'Rating','No. of Ratings','Price'])
    # Write data
    csv_writer.writerows(csv_data)

print('Data has been extracted, saved to data.csv, and HTML pages saved in the output folder.')
```
